- pd.DataFrame()

```python
def frminfo(df) :
    display('data - ' , df )
    print()
    print('shape   - ' , df.shape , df.shape[0] , df.shape[1])
    print('size    - ' , df.size )
    print('ndim    - ' , df.ndim)
    print('index   - ' , df.index    , type(df.index))
    print('columns - ' , df.columns  , type( df.columns ))
    print('values  - \n' , df.values , type(df.values) )
```

```python
print("dict를 이용한 생성-")
dict_data = {
    "col1" : [1,2,3],
    "col2" : [4,5,6],
    "col3" : [7,8,9],
    "col4" : [10,11,12],
    "col5" : [13,14,15]
}

tmp_df = pd.DataFrame(dict_data)
frminfo(tmp_df)
```

```python
tmp_df = pd.DataFrame(dict_data)
display(tmp_df)
```

```python
print(tmp_df["col1"], type(tmp_df["col1"]))
print()
print(tmp_df["col1"].index, tmp_df["col2"].values)
print()
print(tmp_df["col1"][0], type(tmp_df["col1"][0]))
```


#### url을 이용한 json 파일로부터 데이터를 읽어서 데이터프레임을 만들어 보자[실습]
- https://www.kobis.or.kr/kobisopenapi/homepg/apiservice/searchServiceInfo.do?serviceId=searchDailyBoxOffice

```python
import urllib.request
import json
import pandas as pd

url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=f5eef3421c602c6cb7ea224104795888&targetDt=20120101'
response = urllib.request.urlopen(url)

response_data = response.read()  # 데이터를 한 번 읽고 변수에 저장

print('type - ', type(response_data))

json_data = json.loads(response_data)  # 변수를 json.loads() 함수에 전달

data = json_data['boxOfficeResult']['dailyBoxOfficeList']
print(data)

ranks = []
movies = []
amts = []

for d in data:
    ranks.append(d['rank'])
    movies.append(d['movieNm'])
    amts.append(d['salesAmt'])

data_dict = {
    'rank': ranks,
    'title': movies,
    'amount': amts
}

tmp_frm = pd.DataFrame(data_dict)
print(tmp_frm)
display(tmp_frm)
```
